```python
import json
from datasets import load_dataset, Dataset
```

```python
train = load_dataset("wmt16","de-en", split="train[:50000]")
```

```python
train[:]
```

```python
train_dict = train.to_dict()

with open('train_data.json', 'w') as f:
    json.dump(train_dict, f)
```

```python
with open('train_data.json', 'r') as f:
    train_dict = json.load(f)
```

```python
train = Dataset.from_dict(train_dict)
train[:5]
```

```python
validation = load_dataset("wmt16","de-en",split="validation")
```

```python
validation_dict = validation.to_dict()

with open('validation_data.json', 'w') as f:
    json.dump(validation_dict, f)
```

```python
test = load_dataset("wmt16","de-en",split="test")
```

```python
testing_dict = test.to_dict()

with open('test.json', 'w') as f:
    json.dump(testing_dict, f)
```
