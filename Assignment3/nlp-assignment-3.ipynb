```python
training_data = pd.read_csv('/kaggle/input/assignment-3/train.csv',sep="\t")
training_data.isna().sum()
```

```python
training_data = training_data.dropna()
training_data.isna().sum()
```

```python
lst = training_data.values.tolist()
lst[0]
```

```python
sentence1 = []
sentence2 = []
scores = []
for i in range(len(lst)):
    sentence1.append(lst[i][1])
    sentence2.append(lst[i][2])
    scores.append(lst[i][0])
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer("all-MiniLM-L6-v2")

# Compute embedding for both lists
embeddings1 = model.encode(sentence1, convert_to_tensor=True)
embeddings2 = model.encode(sentence2, convert_to_tensor=True)

# Compute cosine-similarities
cosine_scores = util.cos_sim(embeddings1, embeddings2)
```

```python
for i in range(len(sentence1)):
    print("{} \t\t {} \t\t Score: {:.4f}".format(
        sentence1[i], sentence2[i], cosine_scores[i][i]
    ))
```
